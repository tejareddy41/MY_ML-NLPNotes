```python
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
finaldict = []

def processingFileText(fie):

  def remStopWords(text):                      #removing stop words
    stpW=stopwords.words("english")
    return (set(text).difference(stpW))
  
  def punCharDigt(text):                        #removing puncuations
    return [i for i in text if i.isalpha()]

  def sstem(Voca):                              #stememing words
    stemming = PorterStemmer()
    return [stemming.stem(word) for word in Voca]

  def creaVoca(words):                          #creating vocabularies
    bas = words
    set(words)
    thisdict={}
    for i in words:
      thisdict[i] = bas.count(i)
    finaldict.append(thisdict)


  f = open(fie,"r")  #opening file
  text = f.read()    #getting text
  text = text.lower() #lowering words
  wordToken = word_tokenize(text)  #tokensizing words
  remStp = remStopWords(wordToken)  #removing stop words
  punCh = punCharDigt(remStp)    #removing puncuations
  stemi = sstem(punCh)  #steming words
  creaVoca(stemi)       #creating vocabularies


def searchToken(inp):                          #search Function
  stemming = PorterStemmer()
  inp=stemming.stem(inp)
  m=0
  dictf={}
  for i in finaldict:
    m+=1
    if inp in i:
      dictf[m]=i.get(inp)
  sor=reversed((sorted(dictf.items(), key = lambda kv:(kv[1], kv[0]))))
  for key in sor:
    print("Input String at files and repetion count: ",key)



def getFilesandWork():
  filename = ["001.txt","002.txt","003.txt","004.txt","005.txt","006.txt","007.txt","008.txt","009.txt","010.txt","011.txt","012.txt","013.txt","014.txt","015.txt","016.txt","017.txt","018.txt","019.txt","020.txt","021.txt","022.txt","023.txt","024.txt","025.txt","026.txt","027.txt","028.txt","029.txt","030.txt","031.txt","032.txt","033.txt","034.txt","035.txt","036.txt","037.txt","038.txt","039.txt","040.txt","041.txt","042.txt","043.txt","044.txt","044.txt","045.txt","046.txt","047.txt","048.txt","049.txt","050.txt"]
  for i in filename:
    processingFileText(i)
  

if __name__ == "__main__":
  getFilesandWork()
  print("File Dictornaries\n",finaldict)
  searchToken(input("input string to find matches in the Files: "))
```
